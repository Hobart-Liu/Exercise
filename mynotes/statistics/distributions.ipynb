## Geom

```python
from scipy.stats import geom
prob = 0.2
n = 12
p = np.zeros(n)
q = np.zeros(n)
for k in range(1, n+1):
    p[k-1] = (1-prob)**(k-1)*prob
    q[k-1] = geom.pmf(k, prob)
plt.bar(range(1, 13), p, color='green')
plt.xlim(1, 15)
plt.xlabel('n')
plt.ylabel('p(n)')
plt.title('Frist Success at n')
```

```python
plt.bar(range(1, 13), q, color='red')
plt.xlim(1, 15)
plt.xlabel('n')
plt.ylabel('p(n)')
plt.title('First Success at n')
```

## Poisson

```python
from scipy.stats import poisson
import seaborn as sns
poisson_data = poisson.rvs(mu=30, size=10000)
ax = sns.distplot(poisson_data, kde=False, color='purple', hist_kws={'linewidth':15, 'alpha':1})
ax.set(xlabel='Poisson', ylabel='Frequency')
```

```python
mu, sigma = 0.5, 0.1
s = np.random.normal(mu, sigma, 1000)
count, bins, ignored = plt.hist(s, 20, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * 
        np.exp(-(bins-mu)**2 / (2 * sigma**2)), 
        linewidth=3, color='yellow')
print(count)
print(bins)
print(ignored)
```
