```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(start=-5, stop=5, num=100)
```

```python
# linear

w = 2
y = w*x
plt.plot(x,y)
plt.show()
```

### Sigmoid

$ g(a) = \cfrac{1}{1+e^{(-a)}} = \cfrac{1}{1+exp(-a)}$

$ g'(a) = g(a)(1-g(a)) $


```python
def sigmod(a):
    return 1/(1+np.exp(-a))

y = sigmod(x)
plt.plot(x,y)
plt.show()
```

### tanh

$ g(a) = tanh(a) = \frac{exp(a) - exp(-a)}{exp(a) + exp(-a)} = \frac {exp(2a)-1}{exp(2a)+1} $

$ g'(a) = 1 - g(a)^2 $

```python
def tanh(a):
    return (np.exp(2*a)-1)/(np.exp(2*a)+1)

y = tanh(x)
plt.plot(x, y)
plt.show()
```

### ReLU. Rectified Linear

$g(a) = reclin(a) = max(0,a)$

```python
def ReLU(a):
    return np.maximum(0,a)  

y = ReLU(x)
plt.plot(x,y)
plt.show()
```

```python
def ReLU_fast(a):
    return a*(a > 0)

# note: list doesn't support this function, you have to use array
y = ReLU_fast(np.array(x))
plt.plot(x,y)
plt.show()
```
