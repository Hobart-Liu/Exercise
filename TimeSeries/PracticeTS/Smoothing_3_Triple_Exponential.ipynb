We can use Holt Winters to make a forecast on the data set sunspots.: Monthly mean relative sunspot numbers from 1749 to 1983. Collected at Swiss Federal Observatory, Zurich until 1960, then Tokyo Astronomical Observatory. A time series of monthly data from 1749 to 1983. Andrews, D. F. and Herzberg, A. M. (1985) Data: A Collection of Problems from Many Fields for the Student and Research Worker. New York: Springer-Verlag.

Here is a plot: notice that there is not much apparent trend, but a strong seasonality. 

```r
plot(sunspots)
```

Use the forecast equation

$$\hat x_{n+h}=level_n+h∗trend_n+seasonal_{n+h−m} $$

to forecast the activity level one month after data collection has stopped. That is, find $\hat{x}_{n+1} = \hat{x}_{2821}$. You may use the call 

```r
model<- HoltWinters(sunspots)
```

```r
summary(model)
```

```r
model$coefficients
```

```r
level = model$coefficients[1]
trend = model$coefficients[2]
seasonal = model$coefficients[3]
```

```r
level + trend + seasonal
```

We have $ \hat{x}_{n+h} = level_n + h*trend_n + seasonal_{n+h-m}$  

So, $ \hat{x}_{2820+1} = level_{2820} + 1*trend_{2820} + seasonal_{2820+1-12}$  

So, $ \hat{x}_{2821} = 44.44311347 + 1*-0.01320934 + -12.28749771=32.14241$.
