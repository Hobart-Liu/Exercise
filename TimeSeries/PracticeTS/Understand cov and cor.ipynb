## Variance, Standard Deviation, Covariance & Correlation Study Note

__A short summary:__

Variance: single random variable  
variance  
$$\sigma^2 \equiv \frac{\sum{(x-\mu)^2}}{n-1}$$  
standard deviation  
$$\sigma \equiv \sqrt{\frac{\sum{(x-\mu)^2}}{n-1}}$$  

Covariance: 读数没有边界，仅能判定是正相关还是负相关，抑或不相关    
$$cov(x,y) \equiv \frac{\sum{(x- \bar x)(y-\bar y)}}{n-1}$$  

Correlation: [-1, 1]   
$$ cor(x, y) \equiv \frac{cov(x,y)}{\sqrt{variance(x)}\sqrt{variance(y)}} $$

```r
library(ggplot2)
```

随机产生两组数据,  
$y1 = 4 \times x1 + N(0,25)$   
$y2 = 2 \times x2 + N(0, 400)$

比较cov/cor  
cov(x1, y1) vs cov(x2, y2)
cor(x1, y1) vs cor(x2, y2)  

目标：由于y2引入了更多的噪音，希望看到cor(x1, y1) < cor(x2, y2), 但是cov(x1, y1) 和cov(x2, y2) 的比较应该没有意义

```r
len = 100
x1 <- runif(n=len, min=20, max=50)
r1 <- rnorm(n=len, mean=0, sd=5)
y1 <- 4*x1 + r1 - 70
x2 <- runif(n=len, min=20, max=50)
r2 <- rnorm(n=len, mean=0, sd=20)
y2 <- 2 * x2 + r2
ggplot() +
    geom_point(aes(x=x1, y=y1, col='y1')) +
    geom_point(aes(x=x2, y=y2, col='y2')) +
    theme(legend.text=element_text(size=20))
```

验证计算公式

```r
print(sum((x1-mean(x1))^2)/(length(x1)-1))
print(var(x1))
print(sd(x1))
print(sd(x1)^2)
print(cov(x1, x1))
```

```r
print(cov(x1, y1))
print(cov(x2, y2))
```

可以看到，cov(x2, y2) 比 cov(x1, y1)小

```r
print( cov(x1, y1)/(sd(x1)*sd(y1)) )
print( cov(x2, y2)/(sd(x2)*sd(y2)) )
```

相应的 cor(x1,y1) 则远大于 cor(x2, y2)

```r
print(cor(x1, y1))
print(cor(x2, y2))
```
